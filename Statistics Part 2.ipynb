Q1. What is hypothesis testing in statistics?

A1. Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data. It involves making an initial assumption, called the null hypothesis (H₀), and testing it against an alternative hypothesis Ha.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

A2. The null hypothesis (H₀) is a statement that there is no effect or no difference, and it assumes things stay the same. The alternative hypothesis (H₁) is what you want to prove, showing that there is an effect or a difference. In testing, we try to reject the null to support the alternative.

Q3. What is the significance level in hypothesis testing, and why is it important?

A3. The significance level (usually written as α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It sets the threshold for decision-making in hypothesis testing and helps control the risk of making a wrong conclusion (Type I error).

Q4. What does a P-value represent in hypothesis testing?

A4. The P-value shows the probability of getting the observed result, or something more extreme, if the null hypothesis is true. A small P-value (usually less than 0.05) means the result is unlikely under the null hypothesis, so we may reject it.

Q5. How do you interpret the P-value in hypothesis testing?

A5. If the P-value is less than the significance level (e.g., 0.05), we reject the null hypothesis, meaning the result is statistically significant. If the P-value is greater, we fail to reject the null, meaning there's not enough evidence to prove a difference or effect.

Q6. What are Type 1 and Type 2 errors in hypothesis testing?

A6. A Type 1 error happens when we reject a true null hypothesis — it's a false positive.
A Type 2 error happens when we fail to reject a false null hypothesis — it's a false negative.

Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

A7. In a one-tailed test, we check for an effect in one specific direction (greater than or less than).
In a two-tailed test, we check for an effect in both directions (not equal to).

Q8. What is the Z-test, and when is it used in hypothesis testing?

A8. A Z-test is a statistical test used to check if there is a significant difference between sample and population means when the population standard deviation is known and the sample size is large (n ≥ 30). It is commonly used in comparing means and proportions.

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

A9. The Z-score is calculated using the formula:
Z = (X̄ − μ) / (σ / √n)
where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size.
In hypothesis testing, the Z-score shows how far the sample mean is from the population mean in terms of standard deviations. It helps decide whether to accept or reject the null hypothesis.

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

A10. The T-distribution is a probability distribution used when the sample size is small (n < 30) and the population standard deviation is unknown. It is similar to the normal distribution but has fatter tails. We use the T-distribution instead of the normal distribution in such cases to account for more variability and get more accurate results.

Q11. What is the difference between a Z-test and a T-test?

A11. The Z-test is used when the sample size is large (n ≥ 30) and the population standard deviation is known.
The T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown.

Q12. What is the T-test, and how is it used in hypothesis testing?

A12. The T-test is a statistical test used to compare the means of one or two groups when the sample size is small and the population standard deviation is unknown. It helps determine if the difference between sample means is statistically significant. In hypothesis testing, the T-test is used to accept or reject the null hypothesis based on the calculated t-value and the chosen significance level.

Q13. What is the relationship between Z-test and T-test in hypothesis testing?

A13. The Z-test and T-test are both used to test hypotheses about means, but they differ based on sample size and known or unknown standard deviation. The Z-test is used when the sample is large and the population standard deviation is known, while the T-test is used when the sample is small and the standard deviation is unknown. 

Q14. What is a confidence interval, and how is it used to interpret statistical results?

A14. A confidence interval is a range of values that is likely to contain the true population parameter, such as the mean, with a certain level of confidence (like 95%). It helps interpret statistical results by showing how precise an estimate is and gives a sense of the uncertainty in the data. If the interval is narrow, the estimate is more reliable; if it's wide, there's more uncertainty.

Q15. What is the margin of error, and how does it affect the confidence interval?

A15. The margin of error is the amount added and subtracted from the sample estimate to create a confidence interval. It shows how much the estimate might vary due to sampling. A larger margin of error means a wider confidence interval, showing more uncertainty. A smaller margin gives a narrower interval, meaning more precise results.

Q16.  How is Bayes' Theorem used in statistics, and what is its significance?

A16. Bayes' Theorem is used in statistics to update the probability of an event based on new information or evidence. It combines prior knowledge with new data to find a more accurate probability.

Q17. What is the Chi-square distribution, and when is it used?

A17. The Chi-square distribution is a probability distribution used mainly for categorical data. It is used to test how well observed values match expected values in situations like the Chi-square goodness-of-fit test or test of independence. It is applied when working with frequencies or counts in contingency tables and helps check if variables are related or if distributions follow expected patterns.

Q18. What is the Chi-square goodness of fit test, and how is it applied?

A18. The Chi-square goodness of fit test is used to check whether the observed frequencies in a dataset match the expected frequencies based on a specific distribution. It compares how well the data fits a theoretical model. To apply it, you calculate the Chi-square statistic using the formula:
χ² = Σ[(O − E)² / E],
where O is the observed value and E is the expected value. If the result is significantly large, we reject the null hypothesis, meaning the data does not fit the expected distribution.

Q19. What is the F-distribution, and when is it used in hypothesis testing?

A19. The F-distribution is a probability distribution used mainly to compare two variances. It is applied in ANOVA (Analysis of Variance) and other tests where we check if multiple groups have the same variance. In hypothesis testing, the F-distribution helps determine if the differences between group means are statistically significant by analyzing the ratio of variances.

Q20. What is an ANOVA test, and what are its assumptions?

A20.The ANOVA (Analysis of Variance) test is used to check whether there are significant differences between the means of three or more groups. It compares the variation within groups to the variation between groups.
The main assumptions of ANOVA are:
    Independence of observations.
    Normal distribution of the data in each group.
    Equal variances (homogeneity of variance) across the groups.

Q21. What are the different types of ANOVA tests?

A21. There are three main types of ANOVA tests:
One-Way ANOVA – Used to compare the means of three or more independent groups based on one factor.
Two-Way ANOVA – Used to compare means based on two independent factors, and also checks for any interaction between the two.
Repeated Measures ANOVA – Used when the same subjects are tested multiple times, such as before and after a treatment.

Q22. What is the F-test, and how does it relate to hypothesis testing?

A22. The F-test is a statistical test used to compare two or more variances to see if they are significantly different. It is commonly used in ANOVA to test whether the means of different groups are equal. In hypothesis testing, the F-test helps determine if the variability between groups is greater than the variability within groups, which supports rejecting the null hypothesis when the difference is statistically significant.

```python
# Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret 
# the results

import numpy as np
from statsmodels.stats.weightstats import ztest

sample = [72, 75, 78, 74, 77, 76, 73, 79, 74, 76]
population_mean = 74
z_stat, p_value = ztest(sample, value=population_mean)
print("Z-statistic:", round(z_stat, 2))
print("P-value:", round(p_value, 4))
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

```python
# Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(0)
sample = np.random.normal(loc=102, scale=10, size=50)
population_mean = 100
z_stat, p_value = ztest(sample, value=population_mean)
print("Z-statistic:", round(z_stat, 2))
print("P-value:", round(p_value, 4))
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Standard Normal Curve')
plt.axvline(z_stat, color='red', linestyle='--', label=f'Z = {z_stat:.2f}')
z_critical = norm.ppf(0.975) 
plt.fill_between(x, y, 0, where=(x <= -z_critical) | (x >= z_critical), color='lightcoral', alpha=0.5, label='Rejection Region')
plt.title('Two-Tailed Z-Test Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha / 2)
    x_crit_low = mu0 - z_crit * se
    x_crit_high = mu0 + z_crit * se
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    null_dist = norm(mu0, se)
    alt_dist = norm(mu1, se)
    plt.figure(figsize=(10, 5))
    plt.plot(x, null_dist.pdf(x), label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, alt_dist.pdf(x), label='H₁: μ = {}'.format(mu1), color='green')
    plt.fill_between(x, null_dist.pdf(x), where=(x <= x_crit_low) | (x >= x_crit_high), 
                     color='red', alpha=0.4, label='Type I Error (α)')
    plt.fill_between(x, alt_dist.pdf(x), where=(x > x_crit_low) & (x < x_crit_high), 
                     color='orange', alpha=0.4, label='Type II Error (β)')
    plt.axvline(x_crit_low, color='black', linestyle='--')
    plt.axvline(x_crit_high, color='black', linestyle='--')
    plt.title('Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
visualize_type1_type2_errors()
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

from scipy import stats
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    se = stats.sem(data) 
    margin = se * stats.t.ppf((1 + confidence) / 2, df=n-1)
    lower = mean - margin
    upper = mean + margin
    print(f"Sample Mean: {round(mean, 2)}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({round(lower, 2)}, {round(upper, 2)})")
    return (lower, upper)
sample_data = [78, 85, 82, 88, 90, 84, 83, 87, 86, 80]
confidence_interval(sample_data)
```

```python
# Q10.  Write a Python program to calculate the margin of error for a given confidence level using sample data.

from scipy.stats import t
def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * se
    print(f"Sample Mean: {round(mean, 2)}")
    print(f"Margin of Error ({int(confidence * 100)}%): ±{round(margin_of_error, 2)}")
    return margin_of_error
sample_data = [72, 75, 78, 74, 77, 76, 73, 79, 74, 76]
calculate_margin_of_error(sample_data)
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_disease, sensitivity, false_positive):
    prior_no_disease = 1 - prior_disease
    p_positive = (sensitivity * prior_disease) + (false_positive * prior_no_disease)
    posterior = (sensitivity * prior_disease) / p_positive
    print(f"Probability of having the disease given a positive test: {posterior:.4f}")
    return posterior
prior_disease = 0.01
sensitivity = 0.99
false_positive = 0.05
bayes_theorem(prior_disease, sensitivity, false_positive)
```

```python
# Q16. Implement an F-test using Python to compare the variances of two random samples.

from scipy.stats import f
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))
print("Variance of Sample 1:", round(var1, 2))
print("Variance of Sample 2:", round(var2, 2))
print("F-statistic:", round(f_stat, 4))
print("P-value:", round(p_value, 4))
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
# Q18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

import seaborn as sns
from scipy.stats import f_oneway
np.random.seed(1)
group_A = np.random.normal(loc=60, scale=5, size=30)
group_B = np.random.normal(loc=65, scale=5, size=30)
group_C = np.random.normal(loc=70, scale=5, size=30)
f_stat, p_value = f_oneway(group_A, group_B, group_C)
print("F-statistic:", round(f_stat, 2))
print("P-value:", round(p_value, 4))
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")
data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30
})
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=data, palette='pastel')
plt.title('Group Comparison using One-Way ANOVA')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

from scipy.stats import shapiro, levene
def check_anova_assumptions(*groups):
    print("=== Checking Assumptions for ANOVA ===")
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        print(f"\nGroup {i} Normality Test (Shapiro-Wilk):")
        print(f"  W = {stat:.3f}, p = {p:.4f}")
        if p > 0.05:
            print("Likely normal (fail to reject H0)")
        else:
            print("  ❌ Not normal (reject H0)")
    stat, p = levene(*groups)
    print("\nEqual Variance Test (Levene’s):")
    print(f"  W = {stat:.3f}, p = {p:.4f}")
    if p > 0.05:
        print("Equal variances assumed (fail to reject H0)")
    else:
        print("Variances are unequal (reject H0)")
    print("\nIndependence Assumption:")
    print("Must be ensured by study design (e.g., random sampling)")
np.random.seed(1)
g1 = np.random.normal(60, 5, 30)
g2 = np.random.normal(62, 5, 30)
g3 = np.random.normal(65, 5, 30)
check_anova_assumptions(g1, g2, g3)
```

```python
# Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import statsmodels.api as sm
from statsmodels.formula.api import ols
np.random.seed(0)
df = pd.DataFrame({
    'Treatment': np.repeat(['A', 'B'], 15),
    'Diet': np.tile(np.repeat(['Low', 'High'], 15), 1),
    'WeightLoss': np.random.normal(loc=5, scale=1, size=30) +
                  np.repeat([1, 2], 15) +   
                  np.tile(np.repeat([0.5, 1.5], 15), 1)  
})
model = ols('WeightLoss ~ C(Treatment) + C(Diet) + C(Treatment):C(Diet)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("=== Two-Way ANOVA Table ===")
print(anova_table)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Treatment', y='WeightLoss', hue='Diet', data=df, palette='Set2')
plt.title('Two-Way ANOVA: Weight Loss by Treatment and Diet')
plt.xlabel('Treatment Type')
plt.ylabel('Weight Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=25)
sample_variance = np.var(data, ddof=1)
n = len(data)
sigma0_squared = 25 
chi2_stat = (n - 1) * sample_variance / sigma0_squared
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
print(f"Sample Variance: {round(sample_variance, 2)}")
print(f"Chi-square Statistic: {round(chi2_stat, 2)}")
print(f"Critical Region: < {round(chi2_lower, 2)} or > {round(chi2_upper, 2)}")
if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print("Reject the null hypothesis: Population variance is significantly different from σ₀².")
else:
    print("Fail to reject the null hypothesis: No significant difference in population variance.")
```

```python
# Q26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

np.random.seed(1)
data1 = np.random.normal(loc=60, scale=5, size=30)
data2 = np.random.normal(loc=62, scale=8, size=30)
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
if var1 > var2:
    f_stat = var1 / var2
    df1, df2 = len(data1) - 1, len(data2) - 1
else:
    f_stat = var2 / var1
    df1, df2 = len(data2) - 1, len(data1) - 1
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))
print("Variance of Dataset 1:", round(var1, 2))
print("Variance of Dataset 2:", round(var2, 2))
print("F-statistic:", round(f_stat, 3))
print("P-value:", round(p_value, 4))
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
x = np.linspace(0.1, 5, 500)
f_dist = f.pdf(x, df1, df2)
plt.figure(figsize=(8, 5))
plt.plot(x, f_dist, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-statistic = {round(f_stat, 2)}')
plt.fill_between(x, f_dist, where=(x > f.ppf(1 - alpha / 2, df1, df2)) | (x < f.ppf(alpha / 2, df1, df2)),
                 color='red', alpha=0.3, label='Rejection region (α=0.05)')
plt.title('F-test: Variance Comparison')
plt.xlabel('F Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
